```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeat

fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(1,1,1, projection = ccrs.LambertConformal())

ax.add_feature(cfeat.LAND)
ax.add_feature(cfeat.COASTLINE)
ax.add_feature(cfeat.OCEAN)
ax.add_feature(cfeat.BORDERS, linestyle= ':', edgecolor = 'black')
ax.add_feature(cfeat.LAKES, alpha=0.5)
ax.add_feature(cfeat.RIVERS)
ax.gridlines()

ax.set_extent([-5,5,50,55])
```

```python
# Set projection's
proj = ccrs.Stereographic(central_longitude =-95, central_latitude= 35)
proj_nl = ccrs.Orthographic(central_longitude = 5.387638889, central_latitude= 52.156160556) # Amesfoord
```

```python
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(1,1,1, projection = proj_nl)

ax.add_feature(cfeat.LAND)
ax.add_feature(cfeat.COASTLINE)
ax.add_feature(cfeat.NaturalEarthFeature)
ax.add_feature(cfeat.BORDERS, linestyle= ':')
ax.add_feature(cfeat.LAKES, alpha=0.5)
ax.add_feature(cfeat.RIVERS)

ax.set_extend([-5,5,50,55])
```
